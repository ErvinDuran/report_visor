```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Configurar Selenium para cargar la página
driver = webdriver.Chrome()
driver.get("http://192.168.20.38/desarrollo/index.php/rai_report2")

# Obtener el HTML de la página cargada
html = driver.page_source

# Analizar el HTML con BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Buscar la tabla con el ID 'myTable'
table = soup.find('table', {'id': 'myTable'})

# Extraer encabezados automáticamente
if table:
    # Extraer los encabezados (th)
    headers = [th.get_text(strip=True) for th in table.find_all('th')]

    # Extraer los datos de las filas (tr)
    data = []
    for row in table.find_all('tr')[1:]:  # Excluir la fila de encabezados
        cols = row.find_all('td')
        data.append([col.get_text(strip=True) for col in cols])

    # Crear el DataFrame
    df = pd.DataFrame(data, columns=headers)
    print(df)
else:
    print("No se encontró la tabla con el ID 'myTable'.")

# Cerrar el navegador
driver.quit()
```
