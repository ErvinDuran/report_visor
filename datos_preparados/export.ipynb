```python
import pandas as pd
```

```python
df = pd.read_excel('FOTOS.xlsx')
```

```python
def completar(valor):
    valor = str(valor)
    return valor.zfill(2)
```

```python
df['cod_depto']=df['cod_depto'].apply(completar)
df['cod_prov']=df['cod_prov'].apply(completar)
df['cod_mpio']=df['cod_mpio'].apply(completar)
df['concat'] = df['cod_depto']+df['cod_prov']+df['cod_mpio']
```

```python
df = df[['concat','cod_depto', 'depto', 'cod_prov', 'prov', 'cod_mpio', 'mpio']]
```

```python
df.to_csv('municipios.csv',sep=';',index=False)
```
